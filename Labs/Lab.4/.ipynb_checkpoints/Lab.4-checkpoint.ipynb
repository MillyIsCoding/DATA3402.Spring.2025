1. Write a "counter" class that can be incremented up to a specified maximum value, will print an error if an attempt is made to increment beyond that value, and allows reseting the counter. 

```python
class Counter:
    def __init__(self):
        self.counter = 0
        self.maximum = 2
    def Count(self):
        if self.counter < self.maximum:
            self.counter += 1
            print(self.counter)
        else:
            while True:
                reset = input("Maximum Value Reached. Would You Like To Reset The Counter? (y/n) ").strip().lower()
                if reset not in ("y","n"):
                    print("Invalid Input: Try Again")
                elif reset == "y":
                    print("Counter Reset")
                    self.counter = 0
                    break
                else:
                    print("Counter At Max")
                    break
Build = Counter()
print(Build.maximum)
Build.Count()
Build.Count()
Build.Count()
```

4. Implement a class to represent a circle, holding the radius and $x$ and $y$ coordinates of center of the object. Implement functions that compute the area and perimeter of the rectangle. Make all data members private and privide accessors to retrieve values of data members. 

```python
class Circle:
    def __init__(self,radius,center=(0,0)):
        if not isinstance(center, tuple) or len(center) != 2:
            raise ValueError("coord must be a tuple of (x, y)")
        self.__radius = radius
        self.__x,self.__y = center
        self.genPi = 3.14
    def getRadius(self):
        return self.__radius

    def getCoord(self):
        return (self.__x,self.__y)
        
    def getArea(self):
        return (self.genPi*self.__radius)**2
        
    def getCircumference(self):
        return 2*(self.genPi*self.__radius)
circle = Circle(4)
print("Radius:",circle.getRadius())
print("Coordinates:",circle.getCoord())
print("Area:",circle.getArea())
print("Circumference:",circle.getCircumference())
```

7. Add a function to the object classes, including the base, that returns a list of up to 16 pairs of  $x$ and $y$ points on the parameter of the object. 

```python
import math 

class Object:
    def __init__(self,length=None,width=None,radius=None,coord=(0,0)):
        if not isinstance(coord, tuple) or len(coord) != 2:
            raise ValueError("coord must be a tuple of (x, y)")
        self.__length = length
        self.__width = width
        self.__radius = radius
        self.__x,self.__y = coord
        
    def getLength(self):
        if self.__length == None:
            raise NotImplementedError("This Object Has No Length")
        else:
            return self.__length
            
    def getWidth(self):
        if self.__width == None:
            raise NotImplementedError("This Object Has No Width")
        else:
            return self.__width

    def getCoord(self):
        return (self.__x , self.__y)
        
    def getRadius(self):
        if self.__radius == None:
            raise NotImplementedError("This Object Has No Radius")
        else:
            return self.__radius
    
    def getPerimeter(self):
        raise NotImplementedError

    def getArea(self):
        raise NotImplementedError

    def getPoints(self, pairs = 16):
        raise NotImplementedError

class Rectangle(Object):
    def __init__(self,length,width,coord=(0,0)):
        super().__init__(length=length,width=width,coord=coord)
    
    def getArea(self):
        Area = self.getLength() * self.getWidth()
        return Area
        
    def getPerimeter(self):
        Perimeter = (2*(self.getLength() +self.getWidth()))
        return Perimeter

    def getPoints(self,pairs = 16):
        points = []
        length = self.getLength()
        width = self.getWidth()
        x, y = self.getCoord()
        perimeter = self.getPerimeter()
        step = perimeter / pairs
        pos = 0  # Distance traveled along the perimeter
        
        for i in range(pairs):
            if pos < length:
                points.append((x + pos, y))  # Top side
            elif pos < length + width:
                points.append((x + length, y + (pos - length)))  # Right side
            elif pos < 2 * length + width:
                points.append((x + length - (pos - (length + width)), y + width))  # Bottom side
            else:
                points.append((x, y + width - (pos - (2 * length + width))))  # Left side
            pos += step  # Move to the next point

        return points


class Circle(Object):
    def __init__(self,radius,coord=(0,0)):
        super().__init__(radius=radius,coord=coord)
        
    def getArea(self):
        return math.pi*(self.getRadius()**2)
        
    def getPerimeter(self):
        return 2*(math.pi*self.getRadius())
    
    def getPoints(self,pairs=16):
        points = []
        x, y = self.getCoord()
        radius = self.getRadius()
        step = 2 * math.pi / pairs  # Step size in radians

        return [(x + radius * math.cos(step * i), y + radius * math.sin(step * i)) for i in range(pairs)]

class Triangle(Object):
    def __init__(self,length,coord=(0,0)):
        super().__init__(length=length,coord=coord)

    def getArea(self):
        return ((math.sqrt(3)/4) * (self.getLength() **2))
        
    def getPerimeter(self):
        return 3*self.getLength()

    def getVertices(self):
        """ Function Needed to get the coordinates of the pairs """
        x, y = self.getCoord()
        L = self.getLength()

        # First vertex (base left)
        v1 = (x, y)

        # Second vertex (base right)
        v2 = (x + L, y)

        # Third vertex (top)
        height = (math.sqrt(3) / 2) * L
        v3 = (x + L / 2, y + height)
        
        return [v1, v2, v3]
        
    def getPoints(self, pairs = 16):
        vertices = self.getVertices()
        points = []

        for i in range(3):
            x1, y1 = vertices[i]
            x2, y2 = vertices[(i + 1) % 3]  # Connect last vertex to first

            for j in range(pairs // 3):
                t = j / (pairs // 3)
                points.append((x1 + t * (x2 - x1), y1 + t * (y2 - y1)))

        return points
    
# Ex: Triangle
triangle = Triangle(4)
print(f"Triangle Area: {triangle.getArea()}")
print(f"Triangle Perimeter: {triangle.getPerimeter()}")
print(f"Triangle Vertices: {triangle.getVertices()}")
print(f"Triangle Points: {triangle.getPoints()}\n")
# Ex: Rectangle
rect = Rectangle(5, 10)
print(f"Rectangle Area: {rect.getArea()}")       
print(f"Rectangle Perimeter: {rect.getPerimeter()}") 
print(f"Rectangle Points: {rect.getPoints()}\n")

# Ex: Circle
circle = Circle(7)
print(f"Circle Area: {circle.getArea()}")        
print(f"Circle Perimeter: {circle.getPerimeter()}") 
print(f"Circle Points: {circle.getPoints()}")
```

```python
import math 

class Object:
    def __init__(self,length=None,width=None,radius=None,coord=(0,0)):
        if not isinstance(coord, tuple) or len(coord) != 2:
            raise ValueError("coord must be a tuple of (x, y)")
        self.__length = length
        self.__width = width
        self.__radius = radius
        self.__x,self.__y = coord
        
    def getLength(self):
        if self.__length == None:
            raise NotImplementedError("This Object Has No Length")
        else:
            return self.__length
            
    def getWidth(self):
        if self.__width == None:
            raise NotImplementedError("This Object Has No Width")
        else:
            return self.__width

    def getCoord(self):
        return (self.__x , self.__y)
        
    def getRadius(self):
        if self.__radius == None:
            raise NotImplementedError("This Object Has No Radius")
        else:
            return self.__radius
    
    def getPerimeter(self):
        raise NotImplementedError

    def getArea(self):
        raise NotImplementedError

    def getPoints(self, pairs = 16):
        raise NotImplementedError

    def isInside(self,x,y):
        raise NotImplementedError

class Rectangle(Object):
    def __init__(self,length,width,coord=(0,0)):
        super().__init__(length=length,width=width,coord=coord)
    
    def getArea(self):
        Area = self.getLength() * self.getWidth()
        return Area
        
    def getPerimeter(self):
        Perimeter = (2*(self.getLength() +self.getWidth()))
        return Perimeter

    def getPoints(self,pairs = 16):
        points = []
        length = self.getLength()
        width = self.getWidth()
        x, y = self.getCoord()
        perimeter = self.getPerimeter()
        step = perimeter / pairs
        pos = 0  # Distance traveled along the perimeter
        
        for i in range(pairs):
            if pos < length:
                points.append((x + pos, y))  # Top side
            elif pos < length + width:
                points.append((x + length, y + (pos - length)))  # Right side
            elif pos < 2 * length + width:
                points.append((x + length - (pos - (length + width)), y + width))  # Bottom side
            else:
                points.append((x, y + width - (pos - (2 * length + width))))  # Left side
            pos += step  # Move to the next point

        return points
    
    def isInside(self,x,y):
        setX, setY = self.getCoord()
        length = self.getLength()
        width = self.getWidth()
        return setX <= x <= setX + length and setY <= y <= setY + width

class Circle(Object):
    def __init__(self,radius,coord=(0,0)):
        super().__init__(radius=radius,coord=coord)
        
    def getArea(self):
        return math.pi*(self.getRadius()**2)
        
    def getPerimeter(self):
        return 2*(math.pi*self.getRadius())
    
    def getPoints(self,pairs=16):
        points = []
        x, y = self.getCoord()
        radius = self.getRadius()
        step = 2 * math.pi / pairs  # Step size in radians

        return [(x + radius * math.cos(step * i), y + radius * math.sin(step * i)) for i in range(pairs)]

    def isInside(self,x,y):
        setX, setY = self.getCoord()
        rad = self.getRadius()

        return (x - setX) ** 2 + (y - setY) ** 2 <= rad ** 2

class Triangle(Object):
    def __init__(self,length,coord=(0,0)):
        super().__init__(length=length,coord=coord)

    def getArea(self):
        return ((math.sqrt(3)/4) * (self.getLength() **2))
        
    def getPerimeter(self):
        return 3*self.getLength()

    def getVertices(self):
        x, y = self.getCoord()
        L = self.getLength()

        # First vertex (base left)
        v1 = (x, y)

        # Second vertex (base right)
        v2 = (x + L, y)

        # Third vertex (top)
        height = (math.sqrt(3) / 2) * L
        v3 = (x + L / 2, y + height)
        
        return [v1, v2, v3]
        
    def getPoints(self, pairs = 16):
        vertices = self.getVertices()
        points = []

        for i in range(3):
            x1, y1 = vertices[i]
            x2, y2 = vertices[(i + 1) % 3]  # Connect last vertex to first

            for j in range(pairs // 3):
                t = j / (pairs // 3)
                points.append((x1 + t * (x2 - x1), y1 + t * (y2 - y1)))

        return points

    def isInside(self,x,y):
        ###Check if point (x, y) is inside the triangle using the barycentric method.### 
        v1,v2,v3 = self.getVertices()
        
        def area(x1, y1, x2, y2, x3, y3):
            return abs((x1*(y2 - y3) + x2*(y3 - y1) + x3*(y1 - y2)) / 2)

        A = area(*v1, *v2, *v3)  
        A1 = area(x, y, *v2, *v3)  
        A2 = area(*v1, x, y, *v3)  
        A3 = area(*v1, *v2, x, y)  

        return math.isclose(A, A1 + A2 + A3, rel_tol=1e-9)
        
# Ex: Triangle
triangle = Triangle(4)
print(f"Triangle isInside: {triangle.isInside(3,2)}")
# Ex: Rectangle
rect = Rectangle(5, 10)
print(f"Rectangle isInside: {rect.isInside(3,2)}")
# Ex: Circle
circle = Circle(7)
print(f"Circle isInside: {circle.isInside(3,2)}")
```

12. Implement the ability to load/save raster drawings and demonstate that your method works. One way to implement this ability:

   * Overload `__repr__` functions of all objects to return strings of the python code that would construct the object.
   
   * In the save method of raster drawing class, store the representations into the file.
   * Write a loader function that reads the file and uses `eval` to instantiate the object.

For example:

```python
class foo:
    def __init__(self,a,b=None):
        self.a=a
        self.b=b
        
    def __repr__(self):
        return "foo("+repr(self.a)+","+repr(self.b)+")"
    
    def save(self,filename):
        f=open(filename,"w")
        f.write(self.__repr__())
        f.close()
        
   
def foo_loader(filename):
    f=open(filename,"r")
    tmp=eval(f.read())
    f.close()
    return tmp
```
